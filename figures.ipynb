```python
from sklearn.metrics import roc_curve
full = np.loadtxt('full.txt')
none = np.loadtxt('none.txt')
```

```python
y_t = pd.read_csv('data_trouser_dress/troudress_train_y.csv')
y_t = y_t.append(y_t, ignore_index = True)
fpr_f, tpr_f, thresholds_f = roc_curve(y_t, full, drop_intermediate=False)
```

```python
y_n = pd.read_csv('data_trouser_dress/troudress_train_y.csv')
fpr_n, tpr_n, thresholds_n = roc_curve(y_n, none, drop_intermediate=False)
```

```python
plt.plot(fpr_f, tpr_f, label = 'with transformation')
plt.plot(fpr_n, tpr_n, label = 'no transformation')
ax = plt.gca()
ax.set_ylim([0.9, 1])
ax.set_xlim([0.0, 0.1])
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('roc curve')
plt.legend()
```
